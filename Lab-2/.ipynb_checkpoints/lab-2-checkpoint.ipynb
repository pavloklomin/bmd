```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel("вісь X")
plt.ylabel("вісь Y")
plt.title("График")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("вісь X")
plt.ylabel("вісь Y")
plt.subplot(1,2,2)
plt.plot(y,x)
plt.xlabel("вісь X")
plt.ylabel("вісь Y")
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y)
axes.set_xlabel("вісь X")
axes.set_ylabel("вісь Y")
axes.set_title("График")
```

```python

fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.2])
axes1.plot(x, y,'r')
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("График1")
axes2.plot(x, y,'b')
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("График2")
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 4)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label ="x**2",color = "y",lw=5,ls="--")
ax.plot(x, x**3, label ="x**3",color = "b",marker ="o", markersize = 10,markerfacecolor = 'r')
ax.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi =100)
```

```python
fig.savefig("grafik.png")
```
