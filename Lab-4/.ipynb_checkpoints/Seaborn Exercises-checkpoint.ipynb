# Exercises

** Recreate the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.**

** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
sns.jointplot(x = "fare", y = "age", data=titanic, kind="kde", color="#4ca8ed")
# саме з параметром kids ="kde" в нас вийде такий графік, але kids ="hex" буде схожий 
```

```python
sns.distplot(titanic['fare'],kde=False,bins = 20)
```

```python
sns.boxplot(x = "pclass", y = "age", data = titanic)
```

```python
sns.barplot(x = "class", y = "fare", data = titanic,hue="alone")
```

```python
sns.barplot(x = "alone", y = "age", data = titanic)
```

```python
sns.stripplot(x = "alive", y = "age", data = titanic,dodge=True,hue="class")
```

```python
plt.subplots(figsize=(6, 4))
sns.heatmap(titanic.corr())
```

```python
g = sns.FacetGrid(data=titanic, col = "sex")
g.map(plt.hist, "fare")
```

```python
g1 = sns.FacetGrid(titanic, col = "class", row="sex")
g1.map(plt.scatter, "age","fare")
```
