```python

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
sns.distplot(flights['passengers'],bins=50)
```

```python
sns.jointplot(x='year',y='passengers',data=flights)
```

```python
sns.pairplot(flights,hue='year')
```

```python
sns.barplot(x='year',y='passengers',data=flights)
```

```python
sns.barplot(x='year',y='passengers',data=flights,estimator=np.median)
```

```python
sns.boxplot(x='month',y='passengers',data=flights,palette='rainbow')
```

```python
sns.stripplot(x='month',y='passengers',data=flights,hue='year')
```

```python
sns.heatmap(flights.corr())
```
